```python
# Fungsi untuk mendapatkan input baru dan membuat prediksi harga rumah
def predict_house_price(area, bedrooms, bathrooms, parking, mainroad, guestroom, basement, hotwaterheating, airconditioning, furnishingstatus):
    # Buat DataFrame dari input baru
    new_data = pd.DataFrame({
        'area': [area],
        'bedrooms': [bedrooms],
        'bathrooms': [bathrooms],
        'parking': [parking],
        'mainroad_no': [1 if mainroad == 'no' else 0],
        'mainroad_yes': [1 if mainroad == 'yes' else 0],
        'guestroom_no': [1 if guestroom == 'no' else 0],
        'guestroom_yes': [1 if guestroom == 'yes' else 0],
        'basement_no': [1 if basement == 'no' else 0],
        'basement_yes': [1 if basement == 'yes' else 0],
        'hotwaterheating_no': [1 if hotwaterheating == 'no' else 0],
        'hotwaterheating_yes': [1 if hotwaterheating == 'yes' else 0],
        'airconditioning_no': [1 if airconditioning == 'no' else 0],
        'airconditioning_yes': [1 if airconditioning == 'yes' else 0],
        'furnishingstatus_furnished': [1 if furnishingstatus == 'furnished' else 0],
        'furnishingstatus_semi-furnished': [1 if furnishingstatus == 'semi-furnished' else 0],
        'furnishingstatus_unfurnished': [1 if furnishingstatus == 'unfurnished' else 0]
    })

    # Standardisasi input baru
    new_data_scaled = scaler.transform(new_data)

    # Buat prediksi harga
    prediction = model.predict(new_data_scaled)

    # Kembalikan hasil prediksi
    return prediction[0]

# Terima input manual dari pengguna
area = float(input("Masukkan luas area (dalam sqft): "))
bedrooms = int(input("Masukkan jumlah kamar tidur: "))
bathrooms = int(input("Masukkan jumlah kamar mandi: "))
parking = int(input("Masukkan jumlah ruang parkir: "))
mainroad = input("Apakah dekat jalan utama? (yes/no): ").strip().lower()
guestroom = input("Apakah ada kamar tamu? (yes/no): ").strip().lower()
basement = input("Apakah ada basement? (yes/no): ").strip().lower()
hotwaterheating = input("Apakah ada pemanas air? (yes/no): ").strip().lower()
airconditioning = input("Apakah ada AC? (yes/no): ").strip().lower()
furnishingstatus = input("Status perabotan (furnished/semi-furnished/unfurnished): ").strip().lower()

# Lakukan prediksi berdasarkan input pengguna
predicted_price = predict_house_price(area, bedrooms, bathrooms, parking, mainroad, guestroom, basement, hotwaterheating, airconditioning, furnishingstatus)
print(f'Predicted House Price: {predicted_price}')
```
