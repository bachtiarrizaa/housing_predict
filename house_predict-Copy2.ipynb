```python
import pandas as pd

house = pd.read_csv('Housing.csv')
```

```python
# membaca informasi dataset
house.info()
```

```python
# menampilkan isi dataset
house.head(5)
```

```python
house.drop(['stories', 'prefarea'], axis=1, inplace=True)
```

```python
house.head(5)
```

```python
house.replace({'yes': 1, 'no': 0}, inplace=True)
```

```python
# Mengonversi kolom ke tipe data yang sesuai
for col in ['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning']:
    house[col] = house[col].astype(int)
```

```python
house.head(5)
```

```python
# Apply One-Hot Encoding to the 'furnishingstatus' column
house_encoded = pd.get_dummies(house, columns=['furnishingstatus'], drop_first=True)

# View the first few rows of the encoded dataset
print(house_encoded.head())
```

```python
print(house_encoded.dtypes)
```

```python
house_encoded.head(10)
```

```python
X = house_encoded[['area', 'bedrooms', 'bathrooms', 'mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'parking', 'furnishingstatus_semi-furnished', 'furnishingstatus_unfurnished']]
y = house_encoded['price']

# memisahkan menjadi data latih dan data uji
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)
```
